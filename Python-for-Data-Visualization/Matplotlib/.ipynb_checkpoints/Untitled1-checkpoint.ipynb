```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.array([1,2,3,4])
y = np.array([1,2,3,4])
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)
for ax in axes:
    ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)
axes[0].plot(x,y)
```

```python
fig = plt.figure(figsize=(3,2), dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
#OR
fig,axes = plt.subplots(nrows=1, ncols=1)
axes.plot(x,y,label="X")
axes.plot(x,x*2,label="Y")

axes.legend(loc=(0.1, 0.1))
```

```python
y = np.log(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#143123', linewidth=3, linestyle='--')
```
